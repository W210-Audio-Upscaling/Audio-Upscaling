```python
import os
import pandas as pd
from tqdm import tqdm
from pydub.utils import mediainfo
```

```python
# File locations
working_dir = "/Users/tryu/Programs/Audio-Upscaling/"
folders = os.listdir(os.path.join(working_dir, "data"))
folders.remove('.DS_Store')
```

```python
folders
```

```python
files = []
```

```python
for folder in tqdm(folders):
    audio_files = os.listdir(os.path.join(working_dir, "data",folder))
    for audio_file in audio_files:
        media = mediainfo(os.path.join(working_dir, "data", folder, audio_file))
        try:
            del media["DISPOSITION"]
        except:
            pass
        media["name"] = audio_file
        files.append(media)
```

```python
df = pd.DataFrame(files)
```

```python
df = df.dropna()
```

```python
df[["sample_rate", "channels", "bits_per_sample", "duration", "bit_rate", "size"]] = df[["sample_rate", "channels", "bits_per_sample", "duration", "bit_rate", "size"]].apply(pd.to_numeric)
```

```python
df["size_kb"] = df["size"]/1024
```

```python
def gender(_input):
    if "female" in _input:
        return "female"
    else:
        return "male"
```

```python
df["gender"] = df["filename"].apply(lambda x: gender(x))
```
